**Sales and Profitability by Category and Sub-Category**

```python
category_sales = df.groupby(['Category', 'Sub-Category'])['Sales'].sum().sort_values(ascending=False)
category_profit = df.groupby(['Category', 'Sub-Category'])['Profit'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
category_sales.plot(kind='bar', color='y', label='Total Sales')
category_profit.plot(kind='bar', color='r', label='Total Profit')
plt.xlabel('Category, Sub-Category')
plt.ylabel('Amount (USD)')
plt.title('Top-Selling Categories and Sub-Categories')
plt.xticks(rotation=45)
plt.legend()
plt.show()







```

**Impact of Discounts on Profitability**

```python

discount_profit = df.groupby('Discount')['Profit'].sum()

plt.figure(figsize=(8, 6))
discount_profit.plot(kind='bar', color='purple')
plt.xlabel('Discount')
plt.ylabel('Total Profit (USD)')
plt.title('Impact of Discounts on Profitability')
plt.xticks(rotation=0)
plt.show()
```

**Sales distribution by country and region**

```python

if 'Country' in df.columns and 'Region' in df.columns:
    geo_sales = df.groupby(['Country', 'Region'])['Sales'].sum()

    plt.figure(figsize=(10, 6))
    geo_sales.plot(kind='bar', color='b')
    plt.xlabel('Country, Region')
    plt.ylabel('Total Sales (USD)')
    plt.title('Sales Distribution by Country and Region')
    plt.xticks(rotation=45)
    plt.show()
```

**Customer Segmentation (if applicable, assuming column 'Segment' is present)**

```python
# Step 6: Customer Segmentation (if applicable, assuming column 'Segment' is present)
if 'Segment' in df.columns:
    customer_segment_sales = df.groupby('Segment')['Sales'].sum()

    plt.figure(figsize=(8, 6))
    customer_segment_sales.plot(kind='bar', color='orange')
    plt.xlabel('Customer Segment')
    plt.ylabel('Total Sales (USD)')
    plt.title('Sales by Customer Segment')
    plt.xticks(rotation=0)
    plt.show()
```

**Loss-Making Transactions**

```python
loss_transactions = df[df['Profit'] < 0]

plt.figure(figsize=(10, 6))
plt.scatter(loss_transactions['Sales'], loss_transactions['Profit'], color='r')
plt.xlabel('Sales (USD)')
plt.ylabel('Profit (USD)')
plt.title('Loss-Making Transactions')
plt.show()
```
